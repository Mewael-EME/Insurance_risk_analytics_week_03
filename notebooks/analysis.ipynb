```python
# notebooks/analysis.ipynb

# Cell 1: Imports & Setup
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append(os.path.abspath(".."))
from src.eda import summarization, missing_values, univariate, bivariate, outlier_detection


# Load Data
data_path = "insurance_data.txt"
df = summarization.load_data(data_path)
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Cell 2: Data Summarization

# Descriptive Stats for relevant numerical columns
print("Descriptive Statistics:")
display(df[["TotalPremium", "TotalClaims"]].describe())

# Data types
print("\nData Types and Structure:")
display(df.dtypes)
```

```python
# Cell 3: Data Quality Assessment (Missing Values)
print("Missing Values Report:")
missing_values.check_missing(df)
```

```python
# Cell 4: Univariate Analysis
# Uses your module, assumes univariate.py already creates plots
sys.path.append(os.path.abspath(".."))
univariate.plot_distributions(df)
plt.show() 
```

```python
# Cell 5: Bivariate / Multivariate Analysis

# Correlation heatmap of numerical features
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Monthly change trends of TotalPremium vs TotalClaim as a function of ZipCode
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="TotalPremium", y="TotalClaims", hue="PostalCode", alpha=0.7)
plt.title("TotalPremium vs TotalClaim by PostalCode")
plt.xlabel("Total Premium")
plt.ylabel("Total Claims")
plt.legend(loc="best", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
# Cell 6: Data Comparison Across Geography

# Example: Compare TotalPremium per CoverType and ZipCode
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="CoverType", y="TotalPremium", hue="PostalCode")
plt.title("Total Premium by Cover Type and ZipCode")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Compare TotalPremium per PolicyType and ZipCode
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="CoverType", y="TotalPremium")
'''
Other possible x candidates: Product, VehicleType, CoverCategory, Province and ItemType
sns.boxplot(data=df, x="CoverType", y="TotalPremium")
'''
#sns.boxplot(data=df, x="Product", y="TotalPremium")

plt.title("Total Premium by Policy Type and ZipCode")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cell 7: Outlier Detection

# Uses your outlier module
outliers.plot_boxplots(df)
```

```python
# Cell 8: Creative Visualizations

# 1. Premium vs Claim ratio distribution
df["Premium_Claim_Ratio"] = df["TotalPremium"] / (df["TotalClaims"] + 1)
plt.figure(figsize=(10, 5))
sns.histplot(df["Premium_Claim_Ratio"], bins=30, kde=True, color="skyblue")
plt.title("Distribution of Premium-to-Claim Ratio")
plt.xlabel("Ratio")
plt.ylabel("Frequency")
plt.show()

# 2. Most frequent AutoMake per ZipCode
plt.figure(figsize=(4, 80))  # You might want a taller figure now
top_auto = df.groupby(["PostalCode", "make"]).size().reset_index(name="count")
top_auto = top_auto.sort_values("count", ascending=False).groupby("PostalCode").head(1)
sns.barplot(data=top_auto, y="PostalCode", x="count", hue="make", orient='h')
plt.title("Most Common AutoMake by PostalCode")
plt.tight_layout()
plt.show()
```

```python
# 3. Monthly trend of Total Premiums (assuming 'Month' column exists)
if "Month" in df.columns:
    df["Month"] = pd.to_datetime(df["Month"])
    monthly_prem = df.groupby(df["Month"].dt.to_period("M"))["TotalPremium"].sum()
    monthly_prem.plot(kind="line", marker="o", figsize=(10, 4), title="Monthly Total Premium Trend")
    plt.ylabel("Total Premium")
    plt.xlabel("Month")
    plt.show()
else:
    print("Column 'Month' not found for monthly trend plot.")
```
